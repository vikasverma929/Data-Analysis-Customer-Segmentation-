```python
# Step 1: Define Reference Date (last transaction date)
reference_date = initial_df['InvoiceDate'].max()
print("ðŸ“… Reference Date:", reference_date)
```

```python


# Step 2: Compute RFM metrics
rfm = initial_df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',                                   # Frequency
    'TotalAmount': 'sum'                                      # Monetary
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

print("\nâœ… RFM Table (First 5 rows):")
print(rfm.head())

# Step 3: Handle zero or negative Monetary values (from returns)
rfm = rfm[rfm['Monetary'] > 0]

# Step 4: Check Distribution of RFM Variables
plt.figure(figsize=(14,6))
for i, col in enumerate(['Recency', 'Frequency', 'Monetary']):
    plt.subplot(1, 3, i+1)
    sns.histplot(rfm[col], bins=30, kde=True)
    plt.title(f'{col} Distribution')
plt.tight_layout()
plt.show()

# Step 5: Scaling Data (Normalization)
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])

# Step 6: Finding Optimal K (Elbow Method)
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method - Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.show()

# Step 7: Apply K-Means with chosen K (e.g., 4)
kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

# Step 8: Cluster Summary
rfm_summary = rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(2)

print("\nðŸ“Š Cluster Summary:")
print(rfm_summary)

# Step 9: Visualize Clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=rfm, 
    x='Recency', 
    y='Monetary', 
    hue='Cluster', 
    palette='viridis', 
    alpha=0.7
)
plt.title("Customer Segments (Recency vs Monetary)")
plt.show()

# Step 10: Cluster Characteristics
print("\nðŸ§© Cluster Insights:")
for c in rfm['Cluster'].unique():
    print(f"\nCluster {c}:")
    cluster_data = rfm[rfm['Cluster'] == c]
    print(f"ðŸ• Avg Recency: {cluster_data['Recency'].mean():.2f} days")
    print(f"ðŸ” Avg Frequency: {cluster_data['Frequency'].mean():.2f} purchases")
    print(f"ðŸ’° Avg Monetary: Â£{cluster_data['Monetary'].mean():.2f}")

# Step 11: Label Cluster Meanings (Optional)
def label_cluster(row):
    if row['Cluster'] == 0:
        return 'Loyal / High Spenders'
    elif row['Cluster'] == 1:
        return 'New Customers'
    elif row['Cluster'] == 2:
        return 'At Risk / Inactive'
    else:
        return 'Low Value Buyers'

rfm['Segment'] = rfm.apply(label_cluster, axis=1)

print("\nðŸ§¾ Final Segmented Data (First 10 rows):")
print(rfm.head(10))

# Step 12: Save Results for Dashboard or BI Tools
rfm.to_csv('Customer_Segments.csv', index=False)
print("\nðŸ’¾ Segmentation results saved as 'Customer_Segments.csv'")
```
